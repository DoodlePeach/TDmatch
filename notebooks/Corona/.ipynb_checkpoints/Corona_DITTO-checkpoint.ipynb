# Train

```python
import random
import re



file = open('data/er_magellan/Structured/Beer/train.txt','w')

for claim in [m for m in ground_truth.keys()][0:int(0.6*len(ground_truth))]:
#for claim in [m for m in user_claims.keys()][0:int(0.5*len(user_claims))]:

    row = []
    for r in ground_truth[claim]:
        rev = re.sub('\t',' ',r)    
        rev = re.sub('\n',' ',rev)
        rev = re.sub(' +',' ',rev)

        file.write('COL movie VAL' + claim + '\tCOL movie VAL' + rev + '\t1\n' )
        
    for r in random.sample(all_tables,len(all_tables)):
        if ' '.join(r) not in ground_truth[claim]:
            rev = re.sub('\t',' ',' '.join(r))  
            rev = re.sub('\n',' ',rev)
            rev = re.sub(' +',' ',rev)

            file.write('COL movie VAL ' + claim + '\tCOL movie VAL ' + rev + '\t0\n' )


file = open('data/er_magellan/Structured/Beer/valid.txt','w')

for claim in [m for m in ground_truth.keys()][int(0.5*len(ground_truth)):int(0.6*len(ground_truth))]:
#for claim in [m for m in user_claims.keys()][int(0.5*len(user_claims)):int(0.6*len(user_claims))]:

    row = []
    for r in ground_truth[claim]:
        rev = re.sub('\t',' ',r)    
        rev = re.sub('\n',' ',rev)
        rev = re.sub(' +',' ',rev)


        file.write('COL movie VAL' + claim + '\tCOL movie VAL' + rev + '\t1\n' )
        
    for r in random.sample(all_tables,len(all_tables)):
        if ' '.join(r) not in ground_truth[claim]:
            rev = re.sub('\t',' ',' '.join(r))  
            rev = re.sub('\n',' ',rev)
            rev = re.sub(' +',' ',rev)


            file.write('COL movie VAL ' + claim + '\tCOL movie VAL ' + rev + '\t0\n' )



file = open('data/er_magellan/Structured/Beer/test.txt','w')

for claim in [m for m in ground_truth.keys()][int(0.7*len(ground_truth)):int(1.0*len(ground_truth))]:
#for claim in [m for m in user_claims.keys()][int(0.7*len(user_claims)):int(1.0*len(user_claims))]:

    row = []
    for r in ground_truth[claim]:
        rev = re.sub('\t',' ',r)    
        rev = re.sub('\n',' ',rev)
        rev = re.sub(' +',' ',rev)


        file.write('COL movie VAL' + claim + '\tCOL movie VAL' + rev + '\t1\n' )
        
    for r in random.sample(all_tables,len(all_tables)):
        if ' '.join(r) not in ground_truth[claim]:
            rev = re.sub('\t',' ',' '.join(r))  
            rev = re.sub('\n',' ',rev)
            rev = re.sub(' +',' ',rev)

            rev = ' '.join([r for r in rev.split() if not r.isdigit()])

            file.write('COL movie VAL ' + claim + '\tCOL movie VAL ' + rev + '\t0\n' )
```

```python
import re,json,random
file = open('input/corona_test.text','w')


for claim in [m for m in ground_truth.keys()][int(0.5*len(ground_truth)):int(1.0*len(ground_truth))]:
#for claim in [m for m in user_claims.keys()][int(0.5*len(user_claims)):int(1.0*len(user_claims))]:

    row = []
    for r in ground_truth[claim]:
        rev = re.sub('\t',' ',r)    
        rev = re.sub('\n',' ',rev)
        rev = re.sub(' +',' ',rev)

        l = [{"title" : claim}, {"title" : rev}]
        file.write(json.dumps(l) + '\n')        

    for r in random.sample(all_tables,int(len(all_tables))):
        if ' '.join(r) not in ground_truth[claim]:
            rev = re.sub('\t',' ',' '.join(r))  
            rev = re.sub('\n',' ',rev)
            rev = re.sub(' +',' ',rev)

            l = [{"title" : claim}, {"title" : rev}]
            file.write(json.dumps(l) + '\n')
```

```python
from tqdm import tqdm 
file = open('output/corona_output.jsonl','r')

lines = []
for r in file.readlines():
  lines.append(r)
```

```python
from tqdm import tqdm 

ditto_corona = {}

for l in tqdm(lines):
    ll = json.loads(l)
    claim = ll['left']['title']

    if claim not in ditto_corona:
      ditto_corona[claim] = {} 

    rev = ll['right']['title']

        
    if ll['match'] == '0': 
        score = 1 - ll['match_confidence']
    else:
        score = ll['match_confidence']

    ditto_corona[claim][rev] = score
```

```python
for KK in [1,5,20,5000]: 
    i = 0
    MAP, MR, hasP = 0,0,0

    for claim in ditto_corona:
        if claim not in ground_truth or len(ground_truth[claim])==0: continue
        if claim not in user_claims: continue
        
        i+=1
        preds =  [f for (f,j) in   sorted(ditto_corona[claim].items(), key=lambda x: x[1],reverse=True)  ][0:KK]
        golds = [f for f in ground_truth[claim]]

        MAP += MAP_K(golds,preds)
        MR += MRR(golds,preds)
        hasP += HAS_POSITIVE(golds,preds)
        
    print('\n#################### ' + str(KK) + ' ###########################\n')
    print('MRR:',MR/i,'MAP:',MAP/i, 'HAS POSITIVE:', hasP/i)
```
